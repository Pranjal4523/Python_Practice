#### String Manipulation

```python
x="Hello World From Python"
```

```python
type(x)
```

```python
len(x)
```

```python
# Indexing
print(x[0])
print(x[1])
```

```python
print(x[-1])
```

```python
#Slicing
print(x[1:5])  # prints characters from index 1 to 4 (5 is excluded)
print(x[0:8]) # prints characters from index 0 to 7 (8 is excluded)
```

```python
x.count("o") # counts number of occurrences of "o" in the string x
```

```python
x.count("Python") # counts number of occurrences of "Python" in the string x
```

```python
x.upper() # converts all characters in string x to uppercase
```

```python
x.lower() # converts all characters in string x to lowercase
```

```python
x.zfill(25) # increase  the string x with zeros on the left to make its length 25
```

```python
x.ljust(25,'0') # increase  the string x with zeros on the right to make its length 25
```

```python
x.split(" ") # splits the string x at each space and returns a list of substrings
```

```python
list(x) # converts the string x into a list of its characters
```

```python
x=["hello",'how','are','you','today?']
"_".join(x)  # joins the list of strings x into a single string with "_" as separator
```

```python
" ".join(x)  # joins the list of strings x into a single string with " " as separator
```

```python
y="     hello"
y.strip()  # removes leading and trailing whitespace from string y
```

```python
 "_,".join(x)  # joins the list of strings x into a single string with "_, " as separator
```
