    - while loop - Whenever the exact count of repeatation is not known, rather we want to perform condition based repeatation

```python
print("Python is easy and amazing")
print("Python is easy and amazing")
print("Python is easy and amazing")
print("Python is easy and amazing")
print("Python is easy and amazing")
```

```python
ans=5
for i in range(3):
    g=int(input("Enter the number between 0 to 9:"))
    if(g==ans):
        print("You Won The Jackpot")
        break
    else:
        print("Try Again")
```

```python
ans=5
while True:
    g=int(input("Enter the number between 0 to 9:"))
    if g<0 or g>9:
        print("Invalid input, please enter a number between 0 to 9")
        continue # skip the rest and ask for input again and skip the execution below it
    if(g==ans):
        print("You Won The Jackpot")
        break
    else:
        print("Try Again")
```
