## List Comprehension

```python
mtlist=['DXB','DEL','BOM','CCU','MAA','BLR','DDN']

for c in mtlist:
    if c[0]=='D':
        print(c.lower())
```

```python
mtlist=['DXB','DEL','BOM','CCU','MAA','BLR','DDN']
newList=[]

for c in mtlist:
    if c[0]=='D':
        newList.append(c.lower())
print(newList)
```

##### Data Transformation Template
Template 1
   - [f(x) for x in sequance]

```python
mtlist=['DXB','DEL','BOM','CCU','MAA','BLR','DDN']
newList=[c.lower() for c in mtlist]
print(newList)
```

##### Filtering / filtering + transformation
#### Template 2
    - [f(x) for x in sequance if condition]


```python
mtlist=['DXB','DEL','BOM','CCU','MAA','BLR','DDN']
newList=[c.lower() for c in mtlist if c[0]=='D']
print(newList)
```

##### Conditional transformation
#### Template 3
    - [f(x) if condition else g(x) for x in sequance]

```python
mtlist=['DXB','DEL','BOM','CCU','MAA','BLR','DDN']
newList=[c.lower() if c[0]=='D' else c for c in mtlist]
print(newList)
```

## Dictionary Comprehension

#### Template 1
    - {key:f(value) for key,value in dictionary.items()}

```python
mydict={"james":25,"Jessy":36,"Jonas":85,"Jinny":38}

newdic={k:v+5 for k,v in mydict.items()}
print(newdic)
```

### Template 2
- {key:f(v) for key,value in dictionary.items() if condition}

```python
mydict={"james":25,"Jessy":36,"Jonas":85,"Jinny":38}

newdict={k:v+10 for k,v in mydict.items() if v>30}
print(newdict)
```

### Template 3
- {key:f(v) if condition else g(v) for key,value in dictionary.items()}

```python
mydict={"james":25,"Jessy":36,"Jonas":85,"Jinny":38}

newdict={k:v+5 if v>35 else v+10 for k,v in mydict.items()}
print(newdict)
```
