#### Non-Primitive Data Types

### List
- Used to store collection of values, it supports repeatation of values
- A List supports values of multiple data types
- List is mutable - can be modified
- A list can be created in python by specifying a set of values inside square/upper/box bracket=[]

```python
x=[45,6,12.556,'Hello',True,12,True,45]
print(x)
type(x)
```

```python
# To check total number of elements in the list
len(x)
```

```python
# Indexing - Left to Right - Always starts with 0
print(x[0])  # First element
print(x[3])  # Fourth element
print(x[5])  # Sixth element
print(x[11])  # IndexError: list index out of range
```

```python
# Python supports negative indexing as well - right to left- starts with -1
print(x[-1])  # Last element
print(x[-4])  # Fourth last element
print(x[-8])  # Eighth last element
print(x[-9])  # IndexError: list index out of range
```

```python
# Slicing- Accessing a range of elements from the list
print(x[2:5]) # From index 2 to index 4 (5 is excluded) 
```

```python
print(x[0:4]) # From index 0 to index 3 (4 is excluded)
```

```python
print(x[-6:-1]) # From index -6 to index -2 (-1 is excluded)
# [45, 6, 12.556, 'Hello', True, 12, True, 45]
```

```python
# Modify a list element- List is mutable i.e. its elements can be changed
print(x)
x[3]='Python Programming'
print(x)
```

### Range() Vs SLice() Difference

```python
#Range() is used to generate a sequence of numbers(mostly in loops) 
#range(start,stop,step)

for i in range(1,10,2): # Ya Madhe 1 ha start aahe , 10 ha stop aahe ani 2 ha step aahe
                        # 2 step mhnje 1 ntr 2 step mhnje 3 ntr 2 step mhnje 5 ntr asa chalel
                        # if raange(1,10,3) mhnje 1 ntr 4 ntr 7 ntr asa chalel
                        #coz step  aahe 1 ntr 3 rd step mhnje 4 amd 4 ntr  rd step mhnje 7.
    print(i)
```

```python
#Slice() is used to extract a portion of a sequence like list, tuple, string etc.
#slice(start,stop,step)

data=[10,20,30,40,50,60]
s=slice(1,5,2)  # start=1, stop=5, step=2
                # index ithe 0 pasn start hoil mhnje index 1 la 20 aahes ani end la 5 mhnje index 4 prynt chalel
                # mhnje 20 pasn 50 prynt i.e. index 1 pasn index 4 prynt list bhetel
                # tya mdhe 2 chya step ne gheil mhnje output mdhe 20 ani 40 yeil 
print(data[s])  # Output: [20, 40]

# --- IGNORE ---# is slice(0,5,2) # start=0, stop=5, step=2
                 # index ithe 0 pasn start hoil mhnje index 0 la 10 aahes ani end la 5 mhnje index 4 prynt chalel
                # mhnje 10 pasn 50 prynt i.e. index 0 pasn index 4 prynt list bhetel
                # tya mdhe 2 chya step ne gheil mhnje output mdhe 10,30,50 yeil
```
